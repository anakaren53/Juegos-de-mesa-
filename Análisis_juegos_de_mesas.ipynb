```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Juegos de mesa/Juegos de mesa (Respuestas).csv')
df.head()
```

```python
cols_equipos = [
    '¿Qué equipo estás evaluando?',
    '¿Qué equipo estás evaluando?.1',
    '¿Qué equipo estás evaluando?.2'
]
df['Equipos'] = df[cols_equipos].bfill(axis=1).iloc[:, 0]
```

```python
cruzado = df[['Equipos', 'Innovación', 'Jugabilidad y Diseño', 'Presentación']]
cruzado.head()
```

```python
columnas_calif = ['Innovación', 'Jugabilidad y Diseño', 'Presentación']
resultados = df.groupby('Equipos')[columnas_calif].mean()
resultados['Promedio_General'] = resultados.mean(axis=1)
resultados = resultados.round(2)
resultados = resultados.sort_values('Promedio_General', ascending=False)
resultados
```

```python
display(resultados.style.highlight_max(color='lightgreen', axis=0)
                  .format("{:.2f}"))
```

```python
import plotly.express as px
grafico_df = resultados.reset_index()

# 2. Creamos el gráfico
fig = px.bar(grafico_df,
             x='Equipos',
             y=['Innovación', 'Jugabilidad y Diseño', 'Presentación'],
             barmode='group',
             title='Comparativa de Calificaciones por Equipo',
             labels={'value': 'Puntaje', 'variable': 'Criterio', 'Equipos': 'Equipos'},
             color_discrete_sequence=px.colors.qualitative.Pastel1
             )

# 3. Mejoramos el estilo visual (fondo blanco para que resalten los colores)
fig.update_layout(
    plot_bgcolor='rgba(0,0,0,0)',
    yaxis_gridcolor='lightgray'
)

fig.show()
```
